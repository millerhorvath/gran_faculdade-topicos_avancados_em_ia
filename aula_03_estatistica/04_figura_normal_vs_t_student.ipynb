# Distribuição Normal vs T de Student
Este material apresenta uma comparação visual e conceitual entre a distribuição Normal padrão e a distribuição T de Student com diferentes graus de liberdade.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm, t

# Definição das cores baseadas na paleta do GRAN
COR_AZUL_PRINCIPAL = "#0D134C"
COR_VERMELHO = "#DD303E"
COR_VERDE = "#6CA30B"
COR_TEAL = "#137D82"
COR_BRANCO = "#FFFFFF"

# Configuração dos dados
x = np.linspace(-5, 5, 1000)
y_normal = norm.pdf(x, 0, 1)

# Valores de graus de liberdade
dfs = [3, 10, 50]

# Configuração de estilos
estilos = ['--', '-.', ':'] 
cores_t = [COR_TEAL, COR_VERDE, COR_AZUL_PRINCIPAL]

plt.figure(figsize=(12, 6), facecolor=COR_BRANCO)

# Plot da Distribuição Normal Padrão (Base)
plt.plot(x, y_normal, 
         label='Distribuição Normal', 
         linestyle='-', 
         color=COR_VERMELHO, 
         lw=3, 
         zorder=1)

# Plot das Distribuições T (Sobrepostas)
for df, ls, cor in zip(dfs, estilos, cores_t):
    label_text = f'Distribuição $t$ ($df={df}$)'
        
    plt.plot(x, t.pdf(x, df=df), 
             label=label_text, 
             linestyle=ls, 
             color=cor, 
             lw=3, 
             zorder=2)

# Ajuste de escala dos textos para legibilidade em PDF
axis_label_fontsize = 18
plt.xlabel(f'Desvio Padrão ($\\sigma$)', fontsize=axis_label_fontsize, color=COR_AZUL_PRINCIPAL)
plt.ylabel('Densidade de Probabilidade', fontsize=axis_label_fontsize, color=COR_AZUL_PRINCIPAL)

# AUMENTO DOS TICKS (Números nos eixos)
plt.tick_params(axis='both', which='major', labelsize=12)

# AUMENTO DA LEGENDA
# Definido fontsize=16 para garantir leitura em dispositivos móveis ou impressão
plt.legend(loc='upper right', frameon=True, shadow=True, borderpad=1, fontsize=16)

# Ajustes de layout e grid
plt.grid(True, linestyle=':', alpha=0.4, color=COR_AZUL_PRINCIPAL)
plt.gca().set_facecolor('#F9F9F9')

plt.tight_layout(rect=[0, 0.08, 1, 1])

# Configuração do caminho de salvamento
caminho_arquivo = "../Imagens/aula_03_2-2_4-figura-normal_vs_t.png"

# Salvamento da imagem
plt.savefig(caminho_arquivo, format='png', dpi=300, bbox_inches='tight')

print(f"Figura salva com sucesso em: {caminho_arquivo}")

plt.show()
```
